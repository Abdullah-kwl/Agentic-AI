# How Build Tools

```python
from langchain_core.tools import tool
```

# 3 Ways to build tool

### using tool decorator

```python
@tool
def add_tool(a: int, b: int) -> int:
    """Add two numbers."""
    return a + b    
```

```python
print(add_tool.name)
print(add_tool.description)
print(add_tool.args)
print(add_tool.args_schema.model_json_schema())
```

```python
result = add_tool.invoke({"a": 5, "b": 3})
print("sum is:", result)
```

### using langchain structrual tool

```python
from pydantic import BaseModel, Field

class AddToolInput(BaseModel):
    a: int = Field(required=True, description="The first number to add")
    b: int = Field(required=True, description="The second number to add")
```

```python
def structured_add_tool(a:int, b:int) -> int:
    return a + b
```

```python
from langchain_core.tools import StructuredTool

structured_add_tool = StructuredTool.from_function(
    func=structured_add_tool,
    name="Simple_Add",
    description="Useful for adding two numbers.",
    args_schema=AddToolInput
)
```

```python
print(structured_add_tool.name)
print(structured_add_tool.description)
print(structured_add_tool.args)
print(structured_add_tool.args_schema.model_json_schema())
```

```python
result = structured_add_tool.invoke({"a": 5, "b": 3})
print("sum is:", result)
```

### using BaseTool Class from langchain

```python
from langchain.tools import BaseTool

class AddTool(BaseTool):
    name : str = "Simple_Add"
    description : str = "Useful for adding two numbers."
    args_schema : type[BaseModel] = AddToolInput

    def _run(self, a: int, b: int) -> int:
        return a + b
    

my_add_tool = AddTool()
```

```python
print(my_add_tool.name)
print(my_add_tool.description)
print(my_add_tool.args)
print(my_add_tool.args_schema.model_json_schema())
```

```python
result = my_add_tool.invoke({"a": 5, "b": 3})
print("sum is:", result)
```
